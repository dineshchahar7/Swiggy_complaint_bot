```python
class swiggy_bot:

    def datetime_format(self, dayy, timee):
        if dayy.lower() == 'today':
            dayy = datetime.today().strftime('%Y-%m-%d')
            format_string = "%m/%d/%Y"
            full_date = f'{dayy} {timee}'
            converted_date = datetime.strptime(full_date, format_string)
            return converted_date
        else:
            format_string = "%d %b %Y"  # Using %b for abbreviated month name
            current_year = datetime.today().year
            full_date = f'{dayy} {current_year} {timee}'
            converted_date = datetime.strptime(full_date, format_string)
            return converted_date
    def date_filter(self,date_input):
        more_filter = driver.find_element(By.XPATH, '//*[@id="app"]/div[11]/div[2]/div[3]/div/miwi-issues/div/div[1]/div/div/div/div[1]/div/div[2]/div[1]/div[2]/div[1]')
        more_filter.click()
        time.sleep(2)
        WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.CLASS_NAME, 'css-175oi2r.r-1i6wzkk.r-lrvibr.r-1loqt21.r-1otgn73.r-1awozwy.r-14lw9ot.r-1szdl9c.r-1jkafct.r-rs99b7.r-18u37iz.r-1d7fvdj.r-156q2ks.r-127gp16.r-mzo9nz.r-1xtiow5'))
        )
        date_select = driver.find_element(By.CLASS_NAME, 'css-175oi2r.r-1i6wzkk.r-lrvibr.r-1loqt21.r-1otgn73.r-1awozwy.r-14lw9ot.r-1szdl9c.r-1jkafct.r-rs99b7.r-18u37iz.r-1d7fvdj.r-156q2ks.r-127gp16.r-mzo9nz.r-1xtiow5')
        date_select.click()
        time.sleep(2)
        date_element = driver.find_element(By.XPATH, f"//div[@aria-label='{date_input}']")
        date_element.click()
        date_element.click()
        WebDriverWait(driver,10).until(
            EC.visibility_of_element_located((By.XPATH,'//*[@id="app"]/div[11]/div[2]/div[3]/div/miwi-issues/div/div/div[1]/div/div/div[1]/div/div[2]/div[1]/div[2]/div[1]'))

        )
        more_filter_close = driver.find_element(By.XPATH,'//*[@id="app"]/div[11]/div[2]/div[3]/div/miwi-issues/div/div/div[1]/div/div/div[1]/div/div[2]/div[1]/div[2]/div[1]')
        more_filter_close.click()
        return "Date Filter applied"
    def unsolved(self):
        more_filter = driver.find_element(By.XPATH, '//*[text() = "More Filters"]')
        more_filter.click()
        time.sleep(2)
        unresolvedd = driver.find_element(By.XPATH,'//*[text() = "Unresolved"]')
        unresolvedd.click()
        return "Unresolved Filter Applied"
        
    def extract_data(self,local_xpath):
        d = {
            # 'Status' : None,
            'Unique_id' : '',
            # 'Complaint_type' : None,
            'Order_date' : None,
            'Store Name':'',
            'Store_id' : None,
            # 'Store_location' : None,
            # 'Items' : None,
            # 'Customer_Name' : None,
            # 'Customer_history' : None,
            'Order_Date':'',
            'order_id':None,
            'Rating' : None,
            'Comments' : None
        }  
        element = driver.find_element(By.XPATH,local_xpath)
        lines = element.text.splitlines()
        # d['Status'] = lines[0].strip()
        d['Order_date'] = self.datetime_format(lines[1].strip(),lines[2].strip())
        # d['Complaint_type'] = (''.join(lines[3].strip().split('|')[1:])).strip()
        # d['Items'] = lines[4].strip()
        d['Store_id'] = list(map(str,lines[6].strip().split()))[0]
        # d['Store_location'] = list(map(str,lines[6].strip().split()))[1]
        d['Rating'] = lines[8].strip()
        element.click()
        new_element = WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.XPATH,'//*[@id="app"]/div[11]/div[2]/div[3]/div/miwi-issues/div/div[2]/div[1]/div/div/div[2]')))
        newlines = new_element.text.splitlines()
        # d['Complaint_Description'] = newlines[11].strip()
        # d['Customer_Name'] = ' '.join(newlines[7].split())
        d['order_id'] = int(newlines[0].strip()[1:]+newlines[1])
        # d['Customer_history'] = newlines[8].strip() +' '+ newlines[9].strip()+' '+ newlines[10].strip()
        comments = driver.find_element(By.XPATH,'//*[@id="app"]/div[11]/div[2]/div[3]/div/miwi-issues/div/div[2]/div[1]/div/div/div[2]/div[2]/div/div/div/div[8]/div')
        d['Comments'] = comments.text
        return d
    def complaint_close(self):
        view_other_option = driver.find_element(By.XPATH,'//*[text() = "VIEW OTHER OPTIONS"]')
        view_other_option.click()
        time.sleep(1)
        offer_no_resolution = driver.find_element(By.XPATH,'//*[text()= "Offer no resolution"]')
        offer_no_resolution.click()
        time.sleep(1)
        confirmm = driver.find_element(By.XPATH,'//*[text() = "Confirm"]')
        confirmm.click()
        time.sleep(1)
        loosethiscustomer = driver.find_element(By.XPATH,'//*[text() = "Lose this customer"]')
        loosethiscustomer.click()
        time.sleep(1)
        try:
            try:
                try:
                    x1 = '/html/body/div[7]/div/div[2]/div/div/div/div/div/div/div[1]/div/div/img'
                    c1 = driver.find_element(By.XPATH,'/html/body/div[7]/div/div[2]/div/div/div/div/div/div/div[1]/div/div/img')
                    c1.click()
                    time.sleep(2)
                except:
                    tell_us = driver.find_element(By.XPATH,'//*[text() = "Customer issue - item is correct"]')
                    tell_us.click()
                    time.sleep(1)
                    tell_us_confirm = driver.find_element(By.XPATH,'//*[text() = "Confirm"]')
                    tell_us_confirm.click()
                    time.sleep(1)
                    thanku = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div/div/div/div/div/div[1]/div/img')
                    thanku.click()
                    time.sleep(2)                
            except:
                try:
                    try:
                        customer_issue = driver.find_element(By.XPATH,'//*[text() = "Customer issue - item is correct"]')
                        customer_issue.click()
                        time.sleep(1)
                        next1 = driver.find_element(By.XPATH,'//*[text() = "Next (1 of 2)"]')
                        next1.click()
                        time.sleep(1)
                        customer_issue = driver.find_element(By.XPATH,'//*[text() = "Customer issue - item is correct"]')
                        customer_issue.click()
                        confirm2 = driver.find_element(By.XPATH,'//*[text() = "Confirm (2 of 2)"]')
                        confirm2.click()
                        time.sleep(1)
                        thanku = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div/div/div/div/div/div[1]/div/img')
                        thanku.click()
                        time.sleep(2)
                    except:
                        restaurant_packing = driver.find_element(By.XPATH,'//*[text() = "Restaurant Packing/Handover mistake"]')
                        restaurant_packing.click()
                        time.sleep(1)
                        next1 = driver.find_element(By.XPATH,'//*[text() = "Next (1 of 2)"]')
                        next1.click()
                        time.sleep(1)
                        restaurant_packing = driver.find_element(By.XPATH,'//*[text() = "Restaurant Packing/Handover mistake"]')
                        restaurant_packing.click()
                        confirm2 = driver.find_element(By.XPATH,'//*[text() = "Confirm (2 of 2)"]')
                        confirm2.click()
                        time.sleep(1)
                        thanku = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div/div/div/div/div/div[1]/div/img')
                        thanku.click()
                        time.sleep(2)

                except:
                    restaurant_packing = driver.find_element(By.XPATH,'//*[text() = "Restaurant Packing/Handover mistake"]')
                    restaurant_packing.click()
                    confirm = driver.find_element(By.XPATH,'//*[text() = "Confirm"]')
                    confirm.click()
                    time.sleep(1)
                    thanku = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div/div/div/div/div/div[1]/div/img')
                    thanku.click()
                    time.sleep(2)
        except:
            something_went_wrong = driver.find_element(By.XPATH,'//*[text()="Done"]')
            something_went_wrong.click()
            time.sleep(3)
                    
    def complete_bot(self,local_xpath,date_input):
        self.date_filter(date_input)
        self.unsolved()
        time.sleep(4)
        l = []
        while True:
            try:
                l.append(self.extract_data(local_xpath))
                self.complaint_close()
            except:
                break
        time.sleep(4)
        return l
    
       
          
```

```python
def store_id(x):
    x = x.replace(',','')
    x = int(x)
    return x
```

```python
df['Store_id'] = df['Store_id'].apply(store_id)
```

```python
def extract_rating(review):
    match = re.search(r'rated (\d)', review)
    if match:
        return match.group(1)
    else:
        return ''
```

```python
df['Rating'] = df['Rating'].apply(extract_rating)
```
